## Variables in String

We can use the $ 

```julia
#String in String
name = "Charles"
println("Hey, I am $name.")
```

```julia
#Numbers in String; Int and Float
int = 5
float = 4.884
println("My favorite int is $int and my favorite float is $float (kidding i love them all)")
```
